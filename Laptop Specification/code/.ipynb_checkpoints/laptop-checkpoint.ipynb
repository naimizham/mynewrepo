# Laptop Specification Apps

### This application will chose a laptop specification based on company/brand, CPU, GPU, RAM and Memory from user input

```python
import pandas as pd #for reading csv file and create DataFrame
import matplotlib.pyplot as plt #for ploting graph
```

```python
laptop_df = pd.read_csv('../data/laptop_price.csv') #location where csv data have been reading
```

```python
laptop_df.head() #show Dataframe up to 5 rows by default
```

```python
laptop_df['Price_euros'] #to show a series from column 'Price_euros'
```

```python
#create a new 'Price(RM)' column and using lambda function to create new price
laptop_df['Price(RM)'] = laptop_df['Price_euros'].apply(lambda x : x*4.99)
```

```python
laptop_df.isnull().sum() # just for checking any null value
```

```python
#making Laptop class
class Laptop:
    def __init__(self, brand): #object for laptop which are brand/company
        self.brand = brand 
        
    def ram_spec(self,num_ram):#receive input from user and return DataFram specific for RAM only
        return lap_top.list_brand()[lap_top.list_brand()['Ram']==num_ram]
                                    
    def cpu_spec(self,name_cpu):
        return lap_top.list_brand()[lap_top.list_brand()['Cpu']==name_cpu]
                                    
    def gpu_spec(self,name_gpu):
        return lap_top.list_brand()[lap_top.list_brand()['Gpu']==name_gpu]
                                    
    def memory_spec(self,num_memory):
        return lap_top.list_brand()[lap_top.list_brand()['Memory']==num_memory]
                                    
    def cpu_list(self):#show list of cpu
        return lap_top.list_brand()['Cpu'].unique()
    
    def gpu_list(self):
        return lap_top.list_brand()['Gpu'].unique()
    
    def ram_list(self):
        return lap_top.list_brand()['Ram'].unique()
    
    def memory_list(self):
        return lap_top.list_brand()['Memory'].unique()
    
    def list_brand(self):#show the laptop based on company/brand
        return laptop_df[laptop_df['Company'] == self.brand]
    
    def spec(self,cpu,gpu,ram,memory):#receive user input based on Cpu,gpu,ram and memory               
        return laptop_df[(laptop_df['Company'] == self.brand)&(laptop_df['Cpu'] == cpu)&(laptop_df['Gpu'] == gpu)&(laptop_df['Ram'] == ram)&(laptop_df['Memory'] == memory)].head(10)
    
    def expensive_laptop(self):#show top 10 most expensive Laptop from Data
        print('Top 10 Expensive Laptop')
        return laptop_df.groupby(by=['Company']).max().sort_values('Price(RM)',ascending=False).head(10)
    
    def cheap_laptop(self):#show top 10 most cheap Laptop from Data
        print('Top 10 Cheap Laptop')
        return laptop_df.groupby(by=['Company']).min().sort_values('Price(RM)').head(10)
    
    def graph_num_laptop(self):#ploting bar graph of number of laptop for each company
        return laptop_df[['Company']].value_counts().plot(kind='bar', title = 'Number of Laptop for each Company')
    
    def graph_hist_price(self):#ploting histogram bar based on Price
        return laptop_df['Price(RM)'].plot(kind='hist',bins=20, xlabel = 'Price(RM)')
```

```python
lap_top = Laptop(input('Please enter company/brand name:'))#prompt user to enter company/brand name 
lap_top.list_brand()
```
