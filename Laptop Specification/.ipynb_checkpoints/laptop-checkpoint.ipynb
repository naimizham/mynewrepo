```python
laptop_df['Price(RM)'] = laptop_df['Price_euros'].apply(lambda x : x*4.99)
```

```python
class Laptop:
    def __init__(self, brand):
        self.brand = brand
        
    def ram_spec(self,num_ram):
        return lap_top.list_brand()[lap_top.list_brand()['Ram']==num_ram]
                                    
    def cpu_spec(self,name_cpu):
        return lap_top.list_brand()[lap_top.list_brand()['Cpu']==name_cpu]
                                    
    def gpu_spec(self,name_gpu):
        return lap_top.list_brand()[lap_top.list_brand()['Gpu']==name_gpu]
                                    
    def memory_spec(self,num_memory):
        return lap_top.list_brand()[lap_top.list_brand()['Memory']==num_memory]
                                    
    def cpu_list(self):
        return lap_top.list_brand()['Cpu'].unique()
    
    def gpu_list(self):
        return lap_top.list_brand()['Gpu'].unique()
    
    def ram_list(self):
        return lap_top.list_brand()['Ram'].unique()
    
    def memory_list(self):
        return lap_top.list_brand()['Memory'].unique()
    
    def list_brand(self):
        return laptop_df[laptop_df['Company'] == self.brand]
    
    def spec(self,cpu,gpu,ram,memory):               
        return laptop_df[(laptop_df['Company'] == self.brand)&(laptop_df['Cpu'] == cpu)&(laptop_df['Gpu'] == gpu)&(laptop_df['Ram'] == ram)&(laptop_df['Memory'] == memory)].head(10)
    
    def expensive_laptop(self):
        print('Top 10 Expensive Laptop')
        return laptop_df.groupby(by=['Company']).max().sort_values('Price(RM)',ascending=False).head(10)
    
    def cheap_laptop(self):
        print('Top 10 Cheap Laptop')
        return laptop_df.groupby(by=['Company']).min().sort_values('Price(RM)').head(10)
    
    def graph_num_laptop(self):
        return laptop_df[['Company']].value_counts().plot(kind='bar', title = 'Number of Laptop for each Company')
    
    def graph_hist_price(self):
        return laptop_df['Price(RM)'].plot(kind='hist',bins=20, xlabel = 'Price(RM)')
```
